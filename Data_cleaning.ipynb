```python
import pymysql
import pandas as pd

# MySQL connection
myconnection = pymysql.connect(host='127.0.0.1',
                               user='root',
                               passwd='Nikil@8405',
                               database='Data_spark')

# Dictionary mapping CSV filenames to table names
csv_files = {
    "Cleaned_Customers.csv": "Customers",
    "Cleaned_Products.csv": "Products",
    "Cleaned_Sales.csv": "Sales",
    "Cleaned_Stores.csv": "Stores",
    "Cleaned_Exchange.csv": "Exchange",
    "Merged_Dataset.csv": "Merged_Data"
}

# Function to create table and insert data into the database
def upload_csv_to_db(file, table_name):
    try:
        # Load CSV file
        df = pd.read_csv(file)
        
        # Identify columns that should be treated as dates
        date_columns = [col for col in df.columns if 'date' in col.lower() or 'birthday' in col.lower()]
        
        # Convert identified columns to datetime format
        for date_col in date_columns:
            df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
        
        df = df.dropna()  # Drop rows with missing values

        # Replace pandas data types with MySQL-compatible data types
        column = ",".join(
            f"`{col}` {dtype}" 
            for col, dtype in zip(df.columns, df.dtypes)
        ).replace("float64", "FLOAT").replace("object", "TEXT").replace("int64", "INT").replace("datetime64[ns]", "DATE")
        
        # Create table query
        create_table_query = f"CREATE TABLE IF NOT EXISTS `{table_name}` ({column})"
        with myconnection.cursor() as cursor:
            cursor.execute(create_table_query)
        
        # Insert data into the table
        insert_query = f"INSERT INTO `{table_name}` VALUES ({','.join(['%s'] * len(df.columns))})"
        
        # Convert DataFrame rows to list of tuples
        data_to_insert = [tuple(row) for row in df.to_numpy()]
        
        # Insert the data in batch mode
        with myconnection.cursor() as cursor:
            cursor.executemany(insert_query, data_to_insert)
            myconnection.commit()

    except Exception as e:
        print(f"Error uploading data from {file} to {table_name}: {str(e)}")

# Loop through each CSV file and upload it to the database
for file, table_name in csv_files.items():
    upload_csv_to_db(file, table_name)

# Close the MySQL connection
myconnection.close()
```
